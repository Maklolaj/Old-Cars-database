```python


with open('data_cars.txt') as json_file:
    data = json.load(json_file)
    
with open('names.txt') as json_file:
    names = json.load(json_file)

with open('nations.txt') as json_file:
    nations = json.load(json_file)
    
with open('cities.txt') as json_file:
    cities = json.load(json_file)
```

# Functions and DB connection

```python
def create_db_connection(host_name, user_name, user_password, db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password,
            database=db_name
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection

def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query) #usnuąć * przed query?
        connection.commit()
        print("Query successful")
    except Error as err:
        print(f"Error: '{err}'")
    
connection = create_db_connection("localhost", "root", "root", "prodb") # Connect to the Database
```

# Create all tables 

```python
create_brand = """CREATE TABLE IF NOT EXISTS Brand (
idOrigin INT NOT NULL AUTO_INCREMENT,
Name VARCHAR(45) NULL,
PRIMARY KEY (`idOrigin`),
UNIQUE KEY uk_Brand (Name)
);"""
execute_query(connection, create_brand)
```

```python
create_car = """CREATE TABLE IF NOT EXISTS Car (
  idCar INT NOT NULL AUTO_INCREMENT,
  Miles_per_Gallon INT NULL,
  Cylinders INT NULL,
  Displacement INT NULL,
  Horsepower INT NULL,
  Weight_in_lbs INT NULL,
  Acceleration INT NULL,
  Model VARCHAR(45) NOT NULL,
  Year VARCHAR(45) NULL,
  Brand_idBrand INT NOT NULL,
  Origin VARCHAR(45) NULL,
  PRIMARY KEY (idCar),
  FOREIGN KEY (Brand_idBrand) REFERENCES Brand(idOrigin) ON DELETE RESTRICT ON UPDATE CASCADE
  );"""
execute_query(connection, create_car)
```

```python
create_owner ="""CREATE TABLE IF NOT EXISTS Owner (
  idOwner INT NOT NULL AUTO_INCREMENT,
  Name VARCHAR(45) NULL,
  Age INT NULL,
  Nationality VARCHAR(45) NULL,
  PRIMARY KEY (idOwner),
  UNIQUE KEY uk_Owner (Name))
"""
execute_query(connection, create_owner)
```

```python
create_garage = """CREATE TABLE IF NOT EXISTS Garage (
  idGarage INT NOT NULL AUTO_INCREMENT,
  Address VARCHAR(80) NULL,
  Capacity INT NULL,
  Owner_idOwner INT NOT NULL,
  PRIMARY KEY (idGarage),
  UNIQUE KEY uk_Garage (Address),
  FOREIGN KEY (Owner_idOwner) REFERENCES Owner (idOwner) ON DELETE RESTRICT ON UPDATE CASCADE
  )
"""
execute_query(connection, create_garage)
```

```python
create_car_owner = """CREATE TABLE IF NOT EXISTS Car_Owner (
  idCar_Owner INT NOT NULL AUTO_INCREMENT,
  Car_idCar INT NOT NULL,
  Owner_idOwner INT NOT NULL,
  PRIMARY KEY (idCar_Owner),
  FOREIGN KEY (Car_idCar) REFERENCES Car (idCar) ON DELETE RESTRICT ON UPDATE CASCADE,
  FOREIGN KEY (Owner_idOwner) REFERENCES Owner (idOwner) ON DELETE RESTRICT ON UPDATE CASCADE
  )
"""
execute_query(connection, create_car_owner)
```

```python
create_garage_car = """CREATE TABLE IF NOT EXISTS Garage_Car (
  idGarage_Car INT NOT NULL AUTO_INCREMENT,
  Garage_idGarage INT NOT NULL,
  Car_idCar INT NOT NULL,
  PRIMARY KEY (idGarage_Car),
  FOREIGN KEY (Garage_idGarage) REFERENCES Garage (idGarage) ON DELETE RESTRICT ON UPDATE CASCADE,
  FOREIGN KEY (Car_idCar) REFERENCES Car (idCar) ON DELETE RESTRICT ON UPDATE CASCADE
  )
"""
execute_query(connection, create_garage_car)
```
